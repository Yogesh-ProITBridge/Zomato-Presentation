***The data is designed to help researchers, data enthusiasts, and analysts gain insights into the restaurant industry, customer preferences, and regional culinary trends.***

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
z_data = z_data.drop(labels= ['Unnamed: 0.1','Unnamed: 0'],axis=1)
```

```python
z_data.dropna(inplace=True)
```

## Number of restaurants found in different areas?

```python
plt.figure(figsize = (15,13))
sns.countplot(z_data['area'])
plt.title("Number of Restaurants")
plt.xticks(rotation=0)
plt.show
```

## How many restaurants  offer online delivery, and how many do not provide this service?

```python
sns.countplot(data=z_data,x='online_order',y=None)
plt.title("Online and Offline mode count")
plt.show()
```

## online order vs Rate

```python
online_order_counts = z_data['online_order'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(online_order_counts, labels=online_order_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Proportion of Restaurants with Online Orders')
plt.show()
```

```python
df1=pd.crosstab(index=z_data.area,columns=z_data.online_order,margins=True)
df1
```

```python
df1.plot(kind = 'bar',figsize = (15,8))
plt.show()
```

```python
plt.figure(figsize =(8,6))
sns.boxplot(x = 'online_order', y = 'rate (out of 5)', data = z_data)
plt.title("Online Order vs Rating")
plt.show()
```

```python
sns.countplot(data=z_data,x='table booking',y=None)
plt.title("Table Booking Availability")
plt.show()
```

```python
plt.figure(figsize =(8,6))
sns.boxplot(x = 'table booking', y = 'rate (out of 5)', data = z_data)
plt.title("Table Booking vs Rating")
plt.show()
```

```python
correlation_data = z_data[['rate (out of 5)', 'num of ratings', 'avg cost (two people)']]
correlation_matrix = correlation_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)

plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(z_data['num of ratings'], z_data['rate (out of 5)'])
plt.title('Scatter Plot: Ratings vs Number of Ratings')
plt.xlabel('Number of Ratings')
plt.ylabel('Rating (out of 5)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(z_data['avg cost (two people)'], z_data['rate (out of 5)'], c=z_data['num of ratings'], cmap='viridis')
cbar = plt.colorbar(scatter)
cbar.set_label('Number of Ratings')
plt.title('Bubble Chart: Ratings vs Average Cost with Bubble Size representing Number of Ratings')
plt.xlabel('Average Cost (Two People)')
plt.ylabel('Rating (out of 5)')
plt.show()
```

```python
df_sorted = z_data.sort_values(by='num of ratings', ascending=False)
top_three = df_sorted.head(3)
plt.figure(figsize=(10, 6))
plt.bar(top_three['restaurant name'], top_three['num of ratings'], color='blue')
plt.title('Top Three Restaurants Based on Number of Ratings')
plt.xlabel('Restaurant Name')
plt.ylabel('Number of ratings')
plt.xticks(rotation=45, ha='right')
plt.ylim(1000, 20000) 
plt.show()
```

```python
df_sorted = z_data.sort_values(by='rate (out of 5)', ascending=False)
top_two = df_sorted.tail(10)
plt.figure(figsize=(10, 6))
plt.bar(top_two['restaurant name'], top_two['rate (out of 5)'], color='skyblue')
plt.title('Bottom Ten Restaurants Based on Ratings')
plt.xlabel('Restaurant Name')
plt.ylabel('Rating (out of 5)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 5) 
plt.show()
```

```python
df_sorted = z_data.sort_values(by='rate (out of 5)', ascending=False)
top_two = df_sorted.head(10)
new_name = "Spa cuisine"
top_two.iloc[2, top_two.columns.get_loc('restaurant name')] = new_name
plt.figure(figsize=(10, 6))
plt.bar(top_two['restaurant name'], top_two['rate (out of 5)'], color='skyblue')
plt.title('Top Ten Restaurants Based on Ratings')
plt.xlabel('Restaurant Name')
plt.ylabel('Rating (out of 5)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 5) 
plt.show()
```

```python
df_sorted = z_data.sort_values(by='rate (out of 5)', ascending=False)
top_two = df_sorted.tail(10)
plt.figure(figsize=(10, 6))
plt.bar(top_two['restaurant name'], top_two['rate (out of 5)'], color='skyblue')
plt.title('Least Ten Restaurants sorted by their Ratings')
plt.xlabel('Restaurant Name')
plt.ylabel('Rating (out of 5)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 5) 
plt.show()
```
